```
# Install libraries if needed
# !pip install pandas matplotlib seaborn scikit-learn

# Step 0: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Load your dataset
df = pd.read_csv('/content/drive/MyDrive/fc25_cleaned_log_with_league_nation_group.csv')  # ‰øÆÊîπË∑ØÂæÑ

# Step 2: Select numerical features, drop target variables and GK features
numerical_features = df.select_dtypes(include=['int64', 'float64']).drop(columns=[
    'market_value_in_eur', 'log_market_value',
    'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes'
], errors='ignore')

# Step 3: Drop rows with missing values
numerical_features = numerical_features.dropna()

# Step 4: Standardize numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numerical_features)

# Step 5: Apply PCA keeping 8 components
pca = PCA(n_components=8)
pca_features = pca.fit_transform(scaled_features)

# Step 6: Create DataFrame for PCA reduced data
pca_columns = [f'PC{i+1}' for i in range(8)]
df_pca_reduced = pd.DataFrame(pca_features, columns=pca_columns)

# Step 7: Create loadings matrix to explain PCs
loadings = pd.DataFrame(
    pca.components_.T,
    columns=pca_columns,
    index=numerical_features.columns
)

# Step 8: Plot cumulative explained variance
explained_variance = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(10,6))
plt.plot(np.arange(1, len(explained_variance)+1), explained_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance of 8 Principal Components')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 9: Save reduced features and loadings if needed
df_pca_reduced.to_csv('/content/drive/MyDrive/fc25_pca_8components.csv', index=False)
loadings.to_csv('/content/drive/MyDrive/fc25_pca_loadings.csv')

# Step 10: Display results
print("‚úÖ PCA Reduced Dataset Preview:")
display(df_pca_reduced.head())

print("‚úÖ PCA Loadings Matrix Preview:")
display(loadings.head())
```

```
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your cleaned dataset (update your path)
df = pd.read_csv('/content/drive/MyDrive/fc25_cleaned_log_with_league_nation_group.csv')  # ÊîπÊàê‰Ω†ÁöÑË∑ØÂæÑ

# Step 2: Select only important 27 features
important_features = [
    'ball_control', 'def', 'ovr', 'strength', 'pac', 'age', 'weak_foot',
    'jumping', 'dribbling', 'sliding_tackle', 'rank', 'weight_kg',
    'sprint_speed', 'short_passing', 'interceptions', 'pas', 'height_cm',
    'acceleration', 'curve', 'sho', 'dri', 'phy', 'crossing', 'def_awareness',
    'reactions', 'free_kick_accuracy', 'balance'
]
df = df[important_features]

# Step 3: Drop missing values
df = df.dropna()

# Step 4: Calculate correlation matrix
corr_matrix = df.corr().abs()

# Step 5: Find feature pairs with correlation > 0.85
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > 0.85:
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))

# Step 6: Convert to DataFrame and display
high_corr_df = pd.DataFrame(high_corr_pairs, columns=['Feature 1', 'Feature 2', 'Correlation'])

# Show results
print("‚úÖ Highly correlated feature pairs (correlation > 0.85):")
print(high_corr_df)
```

```
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your dataset
df = pd.read_csv('/content/drive/MyDrive/fc25_cleaned_log_with_league_nation_group.csv')  # Êç¢Êàê‰Ω†ÁöÑË∑ØÂæÑÔºÅ

# Step 2: ‰øùÁïôÂéü27‰∏™ÁâπÂæÅ
important_features = [
    'ball_control', 'def', 'ovr', 'strength', 'pac', 'age', 'weak_foot',
    'jumping', 'dribbling', 'sliding_tackle', 'rank', 'weight_kg',
    'sprint_speed', 'short_passing', 'interceptions', 'pas', 'height_cm',
    'acceleration', 'curve', 'sho', 'dri', 'phy', 'crossing', 'def_awareness',
    'reactions', 'free_kick_accuracy', 'balance'
]
df = df[important_features]

# Step 3: Drop missing values
df = df.dropna()

# Step 4: Âà†Èô§10‰∏™ÂÜó‰ΩôÁâπÂæÅÔºåÂè™‰øùÁïô17‰∏™
features_to_drop = [
    'dribbling', 'short_passing', 'curve', 'crossing',
    'sliding_tackle', 'interceptions', 'def_awareness',
    'rank', 'reactions', 'sprint_speed'
]
df_final = df.drop(columns=features_to_drop)

# Step 5: ÁîªÁõ∏ÂÖ≥ÊÄßÁÉ≠ÂäõÂõæÔºàÊ†ºÂ≠êÈáåÊúâÊï∞Â≠óÔºâ
corr_matrix = df_final.corr()

plt.figure(figsize=(16, 14))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5, square=True, cbar_kws={"shrink": 0.75})
plt.title('Correlation Matrix of Final 17 Features (With Values)', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Optional: Check final feature names
print("‚úÖ Final 17 Features:", df_final.columns.tolist())
```

```
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. ËØªÂèñÊï∞ÊçÆ
train_df = pd.read_csv('/content/drive/MyDrive/fc25_train_stratified_by_position.csv')  # ÊîπÊàê‰Ω†ÂÆûÈôÖË∑ØÂæÑ
test_df = pd.read_csv('/content/drive/MyDrive/fc25_test_stratified_by_position.csv')    # Âêå‰∏ä

# 2. ‰∏çÊ†áÂáÜÂåñÁöÑÂàóÔºàÊ†πÊçÆ‰Ω†ÊåáÂÆöÁöÑÈÄªËæëÔºâ
exclude_cols = [
    'position', 'weak_foot', 'skill_moves',
    'market_value_in_eur', 'position_group',
    'log_market_value', 'nation_fifa_ranking',
    'league_ranking_gfl', 'height_m',
    'preferred_foot_binary','pos_Attacker',
    'pos_Defender', 'pos_Goalkeeper', 'pos_Midfielder' ]

# 3. ÊâæÂá∫Ë¶ÅÊ†áÂáÜÂåñÁöÑÁâπÂæÅ
zscore_features = [col for col in train_df.select_dtypes(include='number').columns if col not in exclude_cols]

# 4. Ê†áÂáÜÂåñËÆ≠ÁªÉÈõÜ
scaler = StandardScaler()
train_df_scaled = train_df.copy()
train_df_scaled[zscore_features] = scaler.fit_transform(train_df[zscore_features])

# 5. Ê†áÂáÜÂåñÊµãËØïÈõÜÔºà‰ΩøÁî®Âêå‰∏Ä‰∏™ scalerÔºâ
test_df_scaled = test_df.copy()
test_df_scaled[zscore_features] = scaler.transform(test_df[zscore_features])

# 6. ÂØºÂá∫‰∏∫ CSVÔºàÂ¶ÇÂú® Colab ‰∏≠Ôºâ
train_df_scaled.to_csv('/content/drive/MyDrive/fc25_train_scaled.csv', index=False)
test_df_scaled.to_csv('/content/drive/MyDrive/fc25_test_scaled.csv', index=False)

# 7. ÂèØ‰∏ãËΩΩÊñá‰ª∂
from google.colab import files
files.download('/content/drive/MyDrive/fc25_train_scaled.csv')
files.download('/content/drive/MyDrive/fc25_test_scaled.csv')
```
